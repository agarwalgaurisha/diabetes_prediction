```python
df["Age"]=df["Age"].replace(0,df["Age"].median())
df["DiabetesPedigreeFunction"]=df["DiabetesPedigreeFunction"].replace(0,df["DiabetesPedigreeFunction"].median())
df["Insulin"]=df["Insulin"].replace(0,df["Insulin"].median())
df["SkinThickness"]=df["SkinThickness"].replace(0,df["SkinThickness"].median())
df["BMI"]=df["BMI"].replace(0,df["BMI"].mean())
df["BloodPressure"]=df["BloodPressure"].replace(0,df["BloodPressure"].mean())
df["Glucose"]=df["Glucose"].replace(0,df["Glucose"].mean())
df["Pregnancies"]=df["Pregnancies"].replace(0,df["Pregnancies"].median())

```

```python
df.head()
```

```python
X=df.drop (columns="Outcome",axis=1)
y=df["Outcome"]
```

```python
fig,ax=py.subplots(figsize=(10,10))
sns.boxplot(data=X,ax=ax)
py.savefig("boxplot.jpeg")
```

```python
cols=["Glucose","Pregnancies","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]
for col in cols:
  q1=X[col].quantile(0.25)
  q3=X[col].quantile(0.75)
  iqr=q3-q1
  lower=q1-1.5*iqr
  upper=q3+1.5*iqr
  mask=(X[col]>=lower)&(X[col]<=upper)
```

```python
X_after_outlier_removal=X[mask]
y_after_outlier_removal=y[mask]
```

```python
X_after_outlier_removal.shape
```

```python
y_after_outlier_removal.shape
```

```python

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X_after_outlier_removal)
```

```python
cols=["Glucose","Pregnancies","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]
X_scaled=pd.DataFrame(X_scaled,columns=cols)
```

```python
X_scaled.describe()
```

```python
fig,ax=py.subplots(figsize=(10,10))
sns.boxplot(data=X_scaled,ax=ax)
py.savefig("boxplot.jpeg")
```

```python
y_after_outlier_removal.value_counts()
```

```python
X_scaled.reset_index(drop=True,inplace=True)
y_after_outlier_removal.reset_index(drop=True,inplace=True)
```

```python
q=X_scaled["Insulin"].quantile(0.95)
mask=X_scaled["Insulin"]<q
new_data=X_scaled[mask]
y_after_outlier_removal=y_after_outlier_removal[mask]
```

```python
new_data.shape
```

```python
y_after_outlier_removal.shape
```

```python
fig,ax=py.subplots(figsize=(10,10))
sns.boxplot(data=new_data,ax=ax)
py.savefig("boxplot.jpeg")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(new_data,y_after_outlier_removal,test_size=0.33,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x_train_resample,y_train_resample=smote.fit_resample(x_train,y_train)
print(pd.Series(y_train_resample).value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
classification=LogisticRegression()
classification.fit(x_train_resample,y_train_resample)
```

```python
predictions=classification.predict(x_test)
print(predictions)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

```python
from sklearn.metrics import classification_report
target_names=["diabetics","non_diabetics"]
print(classification_report(y_test,predictions,target_names=target_names))
```
